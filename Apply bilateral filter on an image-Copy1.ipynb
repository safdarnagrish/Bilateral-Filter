## Apply bilateral filter on an image

```python
# Importing Required Packages
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Read the input image

```python
# Read the input image
image = cv.imread('girl.jpg')
```

### Convert Colors

```python
# Convert the image from BGR to RGB for displaying correctly with matplotlib
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

### Display the original image

```python
# Display the original image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
```

### Apply the bilateral filter

```python
# Apply the bilateral filter with adjusted parameters
filtered_image = cv.bilateralFilter(image, d=15, sigmaColor=80, sigmaSpace=80)
```

```python
# Convert the filtered image from BGR to RGB for displaying correctly with matplotlib
filtered_image_rgb = cv.cvtColor(filtered_image, cv.COLOR_BGR2RGB)
```

### Display the filtered image


```python
# Display the filtered image
plt.subplot(1, 2, 2)
plt.imshow(filtered_image_rgb)
plt.title('Filtered Image')
plt.axis('off')

# Show the images
plt.show()
```

# Explain each step in detail 

###  What happened after application of filter? 


The bilateral filter smooths the image while preserving edges. It reduces noise and keeps the edges sharp, unlike other smoothing filters (like Gaussian blur) that might blur the edges.



### What mathematical concept behind that filter?



The bilateral filter combines gray levels or colors based on both their geometric closeness and their photometric similarity. This filter is non-linear and edge-preserving. The output at each pixel is a weighted average of nearby pixels, where the weights depend on both the spatial distance and the intensity difference.

### How that apply to different intesitities of an input image?

The bilateral filter uses both spatial distance and intensity difference to determine the weight of each pixel in the neighborhood. This dual consideration means that pixels that are spatially close and have similar intensity values have a higher influence on the output pixel value. Thus, it preserves edges (where intensity changes sharply) while smoothing regions with similar intensities.
